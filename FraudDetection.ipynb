```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Data = pd.read_csv('Fraud.csv')
```

```python
numeric_data = Data.select_dtypes(include=[np.number])
Q1 = numeric_data.quantile(0.25)
Q3 = numeric_data.quantile(0.75)
IQR = Q3 - Q1
cleaned_data = Data[~((numeric_data < (Q1 - 1.5 * IQR)) | (numeric_data > (Q3 + 1.5 * IQR))).any(axis=1)]
```

```python
Data['isFraud'].value_counts()
```

```python
Normal = Data[Data.isFraud==0]
Fraud = Data[Data.isFraud==1]
```

```python
NormalSample = Normal.sample(n=8213)
```

```python
NormalSample.shape
```

```python
NewData = pd.concat([NormalSample,Fraud],axis=0)
```

```python
NewData.head()
```

```python
NewData['isFraud'].value_counts()
```

```python
x = NewData.drop(['isFraud','step'],axis=1)
y = NewData.isFraud
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
X = pd.get_dummies(NewData, columns=['type','nameOrig','nameDest'])
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,y,test_size=0.2,stratify=y,random_state=123)
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

Training Data Prediction

```python
y_pred_train = model.predict(X_train)
print(f"accuracy: {accuracy_score(Y_train,y_pred_train):.2f}")
```

Testing Data Prediction

```python
y_pred_test = model.predict(X_test)
print(f"accuracy: {accuracy_score(Y_test,y_pred_test):.2f}")
```
